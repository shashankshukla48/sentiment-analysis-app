REMOVE PUNTUATION


```python
#remove puntuations ,.":"? etc
import re
def remove_puntuation(text):
    return re.sub(r'[^a-z0-9\s]', ' ', text).strip()# This function uses a regular expression to find and remove any character that is not a lowercase letter (a-z), a digit (0-9), or a whitespace character.
example_text = "wow... this movie was *amazing*! (10/10). i'd say it's a 'must-see'."
cleaned_text = remove_puntuations(example_text)
print("original", example_text)
print("cleaned", cleaned_text)
```

```python
df['review'] = df['review'].apply(remove_puntuation)
df.head()
```

REMOVE HELPING VERBS AND GRAMMER

```python
#Think about words like: a, the, is, in, on, of, and, i, me, you, he, she, it.
#we will use Natural Language toolkit NLTk
import nltk
nltk.download('stopwords')
```

```python
#we will use set instead of list for all words as time complexity of set is O(1) and list is O(n)
#by default stopword return list so we will convert that into set
from nltk.corpus import stopwords
english_stopword_list = stopwords.words('english')
print("-------First 10 stopwords from NLt liist---------")
print(english_stopword_list[:10])
print("\n"+"="*50+"\n")
```

```python
#convert list into set
stop_words_set = set(english_stopword_list)
print(f"we have successfully loaded {len(stop_words_set)} stopwords into highly efficient set")
print(stop_words_set)
```

Create a Function to Tokenize Text and Remove Stopwords


```python
def tokenization_and_stopwords_remove(text):
    token = text.split()
    cleaned_token = [token for token in token if token not in stop_words_set]
    return cleaned_token
example_text = "this movie is just a brilliant and beautiful production"
cleaned_text = tokenization_and_stopwords_remove(example_text)
print("original", example_text)
print("cleaned", cleaned_text)
```

```python
df['review'] = df['review'].apply(tokenization_and_stopwords_remove)
df.head()
```

LEMMANTIZATION TO TURN WORD TO BACIS FORM

```python
#movies -> movierunning, ran, runs -> run
#Download NLTK WordNet Package for Lemmatization
import nltk
nltk.download('wordnet')
```

```python
#create a funtion to lemmanate and apply to review
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def lemmatize_tokens(tokens):
    lemmatize_tokens = [lemmatizer.lemmatize(tokens) for tokens in tokens]
    return lemmatize_tokens
example_text = "this movie is just a brilliant and beautiful production"
example_text_01 = tokenization_and_stopwords_remove(example_text)
cleaned_text = lemmatize_tokens(example_text_01)
print("original", example_text)
print("cleaned", cleaned_text)
```

```python
df.head()
```

```python
df['review'] = df['review'].apply(lemmatize_tokens)
df.head()
```
